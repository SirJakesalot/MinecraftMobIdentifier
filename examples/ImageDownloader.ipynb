```python
from bs4 import BeautifulSoup
import requests
import re
import urllib.request
import os
import json
```

```python
def get_soup(url,header):
    return BeautifulSoup(urllib.request.urlopen(urllib.request.Request(url,headers=header)),'html.parser')
```

```python
def add_query_links(links, search_url, header):
    soup = get_soup(search_url, header)
    original_len = len(links)
    num_links_found = 0
    for a in soup.find_all('div', {'class':'rg_meta'}):
        link, ext = json.loads(a.text)['ou'], json.loads(a.text)['ity']
        links[link] = ext
        num_links_found += 1
    num_dups = num_links_found - (len(links) - original_len)
    print('Num Links Found: {0}, Num Duplicates Found: {1}'.format(num_links_found, num_dups))
```

```python
queries = ['minecraft', 'minecraft gameplay', 'minecraft screenshots', 'minecraft']
tokenized = ['+'.join(query.split()) for query in queries]
urls=['https://www.google.co.in/search?q={0}&source=lnms&tbm=isch'.format(tokens) for tokens in tokenized]
header={'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.134 Safari/537.36"
}
```

```python
links = dict()
for url in urls:
    add_query_links(links, url, header)
```

```python
save_dir = 'pics'
num_errors = 0

if not os.path.exists(save_dir):
    os.mkdir(save_dir)
    
for i, (link, ext) in enumerate(links.items()):
    try:
        raw_img = urllib.request.urlopen(link).read()
        save_path = os.path.join(save_dir, '{0}.'.format(i))
        if ext not in ['jpg', 'png', 'jpeg']:
            save_path += 'jpg'
        else:
            save_path += ext
        with open(save_path, 'wb') as f:
            f.write(raw_img)
        print('{0} at {1}'.format(save_path, link))
    except Exception as e:
        print('{0} could not load {1}'.format(i, link))
        print(e)
        num_errors += 1

print('number of errors: {0}'.format(num_errors))
```
